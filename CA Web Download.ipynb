```python
#Script used to download '.xls' files 
#This script can also be used to download other formats
from os.path import basename
from urllib.parse import urljoin
from lxml import html
import bs4
import requests

base_url = 'https://www.cde.ca.gov/ds/sp/ai/'
page = requests.get(base_url).text
```

```python
doc = bs4.BeautifulSoup(page,'lxml')
```

```python
print(doc.prettify())
```

```python
at=[]
for links in doc.findAll('a', href=True):
    if(links['href'].endswith('.xls')):
        print(links['href'])  
        at.append(links['href'])
    
```

```python
iit=[]
for i in at:
    if(i.startswith('documents')):
        url = urljoin(base_url, i)
        iit.append(url)
        print(url)
    else:
        print(i)
        iit.append(i)
```

```python
 
for t in iit:   
    with open("C:/Users/Sai Shanoor/" + basename(t), 'wb') as f:
        print("Downloading", t)

        data = requests.get(t).content
        f.write(data) 
```
