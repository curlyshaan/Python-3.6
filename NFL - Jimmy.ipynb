```python
#Data visualisation of Football Scores


import pandas as pd
import numpy as np
import datetime as dt 

import matplotlib.pyplot as plt 
from matplotlib import cm

plt.style.use('ggplot') # use the ggplot plotting 

%matplotlib inline 
from __future__ import division 
```

```python
nfl=pd.read_csv('C:/Data Analysis/pbp_2017.csv', low_memory=False)
```

```python
nfl.head(3)
```

```python
nfl['Date'] = pd.to_datetime(nfl['Date'])
```

```python
nfl['Date'].head(5)
```

```python
nfl['Jimmy'] = np.where( nfl['Date'] >= pd.datetime(2017,12,3), 'yes', 'no')
```

```python
nfl['Jimmy'].head()
```

```python
nfl.head()
```

```python
niners = nfl[ (nfl["HomeTeam"] == 'SF') | (nfl["AwayTeam"] == 'SF') ]
```

```python
niners
```

```python
niners_td = nfl[((nfl["HomeTeam"] == 'SF') | (nfl["AwayTeam"] == 'SF')) & (nfl["sp"] == 1) & (nfl["Touchdown"] == 1) & (nfl["DefensiveTeam"] != 'SF') & pd.isnull(nfl["Interceptor"]) ]
```

```python
niners_td.count()
```

```python
#touchdowns without Jimmy
niners_td.groupby('Jimmy').Touchdown.sum()
```

```python
#we know that the number of touchdowns with jimmy
temp=niners_td.groupby('Jimmy').Touchdown.sum()
temp.plot(kind='bar', alpha=1.0)
plt.title('Distribution of Touch downs')
plt.xlabel('Did Jimmy play the game?')
plt.ylabel('Number of Touch Downs')
plt.yticks(np.arange(0, 20, step=1))
plt.show()
```

```python

#finding how many gaMes jimmy played in
pre_jim=len(niners[niners['Jimmy']=='no']['GameID'].unique())
post_jim=len(niners[niners['Jimmy']=='yes']['GameID'].unique())
```

```python
print("Number of games without Jimmy:", pre_jim)
print("Number of games with Jimmy:", post_jim)
```

```python
#touchdowns by game
td_date = niners.groupby('Date')['Touchdown'].sum()
td_date
```

```python
it_date=niners.groupby('Date').InterceptionThrown.sum()
it_date
```

```python
fig, ax = plt.subplots(figsize=(8, 6), dpi = 80) # set plot size 


f1 = td_date.plot() # plotting the touchdowns
f2 = it_date.plot() # plotting the interceptions

ax.set_title("Touchdowns and Interceptions over Time", fontsize = 18) 
ax.set_xlabel('Game Date') 
ax.set_ylabel('Count')
plt.xticks()
plt.yticks()

plt.axvline(dt.datetime(2017, 12, 3), color = 'black') #vertical line showing Jimmy's join date
plt.legend(loc='upper center', frameon=True, facecolor="yellow") 

plt.show()
```
